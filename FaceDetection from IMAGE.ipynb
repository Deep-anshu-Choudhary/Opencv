```python
# Step 2 making object of Model
model=cv.CascadeClassifier("haar_frontal.xml")   #REQUIREMENT OF THIS mODEL IS TO TAKE IMAGE IN GRAY SCALE
```

```python
#Step3 Convert original source image into grayscale image
img_gray=cv.cvtColor(img_rgb,cv.COLOR_RGB2GRAY)
plt.imshow(img_gray,cmap='gray')
```

```python
# Step Predicts the faces
faces=model.detectMultiScale(img_gray,1.3,5)     # 1.3 is scale factor ie size and 
                                                 #5 is minimun pixels between the two images 
```

```python
faces.shape
```

```python
faces
```

```python
faces[0]
```

```python
x,y,w,h=faces[0]     #(x,y) is top left   and w is width and h is height
```

```python
cv.rectangle(img_rgb,(x,y),(x+w,y+h),(255,255,255),5)
plt.imshow(img_rgb)
```

```python
for i in faces:
    x,y,h,w=i
    cv.rectangle(img_rgb,(x,y),(x+w,y+h),(255,255,255),5)
plt.imshow(img_rgb)
```
